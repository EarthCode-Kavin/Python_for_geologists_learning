```python
from pathlib import Path
import os

DATA = Path("..") / "data"
print("cwd            :", Path.cwd())
print("data dir exists:", DATA.exists())
```

## 1. Paths are objects, not strings
`/` joins paths on **any** operating system — no more `\\` headaches.

```python
f = DATA / "assay.csv"
print("name  :", f.name)
print("stem  :", f.stem)
print("suffix:", f.suffix)
print("parent:", f.parent)
print("size  :", f.stat().st_size, "bytes")
```

## 2. Glob — find every CSV in a project

```python
for csv in sorted(DATA.glob("*.csv")):
    kb = csv.stat().st_size / 1024
    print(f"{csv.name:<20} {kb:8.1f} KB")
```

## 3. Create output folders safely

```python
out = Path("outputs") / "processed"
out.mkdir(parents=True, exist_ok=True)     # no error if it already exists
print("ready:", out.resolve())
```

## 4. Read and write text files

```python
log = out / "run_log.txt"
log.write_text("processed 3 datasets\nno errors\n", encoding="utf-8")
print(log.read_text(encoding="utf-8"))
```

## 5. `os` — environment and system info

```python
print("OS name      :", os.name)
print("CPU count    :", os.cpu_count())
print("HOME/USERPROFILE set:", bool(os.environ.get("HOME") or os.environ.get("USERPROFILE")))
```

## 6. Batch pattern — process every file in a loop

```python
import pandas as pd

for csv in sorted(DATA.glob("*.csv")):
    df = pd.read_csv(csv)
    print(f"{csv.stem:<15} rows={len(df):<6} cols={df.shape[1]}")
```
