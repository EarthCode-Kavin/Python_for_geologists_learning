```python
import numpy as np
print("NumPy", np.__version__)
```

## 1. Arrays vs lists — vectorised math
Operations apply to *every element at once*, no loop needed.

```python
cu_pct = np.array([0.79, 0.83, 1.20, 0.45, 2.10, 0.31, 1.75])

cu_ppm = cu_pct * 10000            # whole-array math
print("ppm:", cu_ppm)
print("mean %.3f | std %.3f | max %.2f" % (cu_pct.mean(), cu_pct.std(), cu_pct.max()))
```

## 2. Boolean masking — apply a cutoff grade

```python
cutoff = 0.5
mask = cu_pct >= cutoff
print("mask        :", mask)
print("ore grades  :", cu_pct[mask])
print("ore fraction: %.0f%%" % (100 * mask.mean()))
print("mean ore grade: %.2f %% Cu" % cu_pct[mask].mean())
```

## 3. Creating arrays — depth profiles and grids

```python
depth = np.arange(0, 100, 2.0)          # 0 to 98 m every 2 m (like assay FROM)
temps = 15 + 0.03 * depth               # geothermal gradient 30 °C/km

print("depth[:5]:", depth[:5])
print("temp @ 98 m: %.2f °C" % temps[-1])

grid = np.zeros((3, 4))                 # 2-D array = the shape of a raster
print("grid shape:", grid.shape)
```

## 4. 2-D arrays — a mini elevation model

```python
rng = np.random.default_rng(42)
dem = 2500 + 300 * rng.random((5, 5))   # 5x5 elevation grid in metres

print(np.round(dem, 0))
print("highest cell :", dem.max().round(1), "m at index", np.unravel_index(dem.argmax(), dem.shape))
print("mean per row :", dem.mean(axis=1).round(1))
```

## 5. Useful everyday functions

```python
au_gpt = np.array([1.75, 1.73, 0.20, np.nan, 3.10, 0.85])

print("nan-safe mean :", np.nanmean(au_gpt).round(3), "g/t")
print("sorted        :", np.sort(au_gpt))
print("cumulative    :", np.nancumsum(au_gpt).round(2))
print("clipped 0-2   :", np.clip(au_gpt, 0, 2))
print("percentile 75 :", np.nanpercentile(au_gpt, 75).round(2))
```
