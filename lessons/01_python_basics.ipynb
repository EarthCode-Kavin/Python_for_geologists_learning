## 1. Variables and types
A rock sample is just data: numbers, text, booleans.

```python
sample_id = "DH001-042"        # str  - text
depth_m   = 84.5               # float - decimal number
cu_pct    = 1.32               # float - copper grade in %
is_ore    = cu_pct > 0.5       # bool  - True/False (0.5% Cu cutoff)

print(f"Sample {sample_id} at {depth_m} m: {cu_pct}% Cu -> ore? {is_ore}")
print(type(sample_id), type(depth_m), type(is_ore))
```

## 2. Lists — ordered collections
A drillhole is a *list* of measurements.

```python
grades = [0.79, 0.83, 1.20, 0.45, 2.10, 0.31, 1.75]   # Cu % down the hole

print("n samples :", len(grades))
print("first/last:", grades[0], grades[-1])
print("top 3     :", grades[:3])          # slicing
print("mean      :", sum(grades) / len(grades))

grades.append(0.95)                        # add a new assay
print("after append:", grades)
```

## 3. Dictionaries — labelled data
Perfect for a single geochemical analysis (the shape Pandas rows take later).

```python
andesite = {"SiO2": 62.7, "Al2O3": 16.9, "CaO": 5.31, "Na2O": 4.45, "K2O": 1.57}

print("SiO2 =", andesite["SiO2"], "wt%")
andesite["MgO"] = 3.39                     # add a key
total = sum(andesite.values())
print(f"sum of oxides so far: {total:.2f} wt%")

for oxide, wt in andesite.items():
    print(f"  {oxide:>6}: {wt} wt%")
```

## 4. Loops and conditionals
Classify earthquake magnitudes the way a seismologist would.

```python
magnitudes = [4.2, 5.6, 7.5, 6.1, 3.8, 8.0]

for m in magnitudes:
    if m >= 7.0:
        label = "major"
    elif m >= 5.0:
        label = "moderate"
    else:
        label = "minor"
    print(f"M{m:.1f} -> {label}")
```

## 5. Functions — reusable geology logic

```python
def classify_sio2(sio2_wtpct):
    """TAS-style silica classification of a volcanic rock."""
    if sio2_wtpct < 52:
        return "basic (basalt family)"
    elif sio2_wtpct < 63:
        return "intermediate (andesite family)"
    else:
        return "acid (dacite/rhyolite family)"

for sio2 in [48.5, 57.2, 62.7, 71.0]:
    print(f"SiO2 {sio2} wt% -> {classify_sio2(sio2)}")
```

## 6. List comprehensions — one-line loops

```python
grades = [0.79, 0.83, 1.20, 0.45, 2.10, 0.31, 1.75]

ore      = [g for g in grades if g >= 0.5]          # filter by cutoff
grams_pt = [g * 10000 for g in grades]              # % -> ppm
print("ore intervals :", ore)
print("grades in ppm :", grams_pt)
```
